```python
import h5py
import numpy as np
import matplotlib
import sys
#matplotlib.use('Agg')
matplotlib.rcParams['mathtext.default'] = 'regular'
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from PIL import Image
from numpy import inf
import numpy.ma as ma
from numpy import *
```

```python
#Loading save number density files (n(v) at different j,k values)
n_v_SiII=np.load('n_v_SiII.npy')
n_v_OVI=np.load('n_v_OVI.npy')
n_v_CII=np.load('n_v_CII.npy')
```

```python
#Constant that are needed to calculate Tau(v)
me=9.10e-28 #mass of electron in grams
c=3e18 #speed of light in Angstroms per sec
dv=1.2e6  #in cm length of velocity 
dr=4.94e20 #length of box in cm length along the x-axis
e = 4.80e-10 #in esu=((cm^3/2)*(grams^1/2))/s
#oscillator strength 
f_s_SiII=1.18
f_w_SiII=0.582
f_s_OVI=0.1325
f_w_OVI=0.0658
f_CII=0.128
#rest wave-length
rw_s_SiII=1260 # in Angstroms
rw_w_SiII=1193
rw_s_OVI=1032 #in Angstroms
rw_w_OVI=1038
rw_CII=1335 #in Angstroms
```

```python
#Calculating T(v) at different j, k values
Tau_jks_SiII= ((((np.pi)*(e**2))/(me*c))*f_s_SiII*n_v_SiII*rw_s_SiII*(dr/dv))
Tau_jkw_SiII= ((((np.pi)*(e**2))/(me*c))*f_w_SiII*n_v_SiII*rw_w_SiII*(dr/dv))

Tau_jks_OVI= ((((np.pi)*(e**2))/(me*c))*f_s_OVI*n_v_OVI*rw_s_OVI*(dr/dv))
Tau_jkw_OVI= ((((np.pi)*(e**2))/(me*c))*f_w_OVI*n_v_OVI*rw_w_OVI*(dr/dv))

Tau_jk_CII= ((((np.pi)*(e**2))/(me*c))*f_CII*n_v_CII*rw_CII*(dr/dv))
```

```python
#Tau_jks_SiII_50= ((((np.pi)*(e**2))/(me*c))*f_s_SiII*n_v_SiII_reshape*rw_s_SiII*(dr/dv))
#Tau_jkw_SiII_50= ((((np.pi)*(e**2))/(me*c))*f_w_SiII*n_v_SiII_reshape*rw_w_SiII*(dr/dv))
```

```python
dx = 2.411e17# size scale, in cm, 5/64 pc
T_jkw_SiII= ((((np.pi)*(e**2))/(me*c))*f_w_SiII*n_v_SiII*rw_w_SiII*(dr/dv))
test=T_jkw_SiII[:,:,4]
test1=np.reshape(test,512*512)
test2=T_jkw_SiII.mean(axis=(0,1))
#plt.plot(test2)
#plt.show()
np.max(test2)
```

```python
#Change this to fabsorbed
f_s_absorbed_SiII=np.exp(-Tau_jks_SiII)
f_w_absorbed_SiII=np.exp(-Tau_jkw_SiII)

f_s_absorbed_OVI=np.exp(-Tau_jks_OVI)
f_w_absorbed_OVI=np.exp(-Tau_jkw_OVI)

f_absorbed_SiII=np.exp(-Tau_jks_SiII)
f_absorbed_CII=np.exp(-Tau_jk_CII)
f_absorbed_OVI=np.exp(-Tau_jks_OVI)
```

```python
#Average fabsorbed
avgf_s_absorbed_SiII=f_s_absorbed_SiII.mean(axis=(0,1))
avgf_w_absorbed_SiII=f_w_absorbed_SiII.mean(axis=(0,1))

avgf_s_absorbed_OVI=f_s_absorbed_OVI.mean(axis=(0,1))
avgf_w_absorbed_OVI=f_w_absorbed_OVI.mean(axis=(0,1))

avgf_absorbed_SiII=f_absorbed_SiII.mean(axis=(0,1))
avgf_absorbed_CII=f_absorbed_CII.mean(axis=(0,1))
avgf_absorbed_OVI=f_absorbed_OVI.mean(axis=(0,1))
```

```python
c_SiII=((avgf_w_absorbed_SiII-1.)**2.)/(avgf_s_absorbed_SiII-1.-(2.*(avgf_w_absorbed_SiII-1.)))
c_OVI=((avgf_w_absorbed_OVI-1.)**2.)/(avgf_s_absorbed_OVI-1.-(2.*(avgf_w_absorbed_OVI-1.)))
```

```python
log_1_Fw_SiII=np.log(c_SiII)+np.log(1-avgf_w_absorbed_SiII)
log_1_Fw_OVI=np.log(c_OVI)+np.log(1-avgf_w_absorbed_OVI)
```

```python
Tau_0_OVI=np.log(1-avgf_w_absorbed_OVI)-np.log(avgf_w_absorbed_OVI-avgf_s_absorbed_OVI)
Tau_0_SiII=np.log(1-avgf_w_absorbed_SiII)-np.log(avgf_w_absorbed_SiII-avgf_s_absorbed_SiII)
```
