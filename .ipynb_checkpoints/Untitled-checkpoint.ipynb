```python
values_Si=np.zeros((ny,nz,100))
for j in range(0,256):
    for k in range(0,256):
        N_Si_II=n_SiII[:,j,k]
        vel_Si=vx[:,j,k]
        values_Si[j,k,:],b=np.histogram(vel_Si,bins=100,weights=N_Si_II,range=(0,1200))
```

```python
values_Si=np.zeros((ny,nz,100))
for j in range(0,256):
    for k in range(0,256):
        N_Si_II=n_SiII[:,j,k]
        vel_Si=vx[:,j,k]
        values_Si[j,k,:],b=np.histogram(vel_Si,bins=100,weights=N_Si_II,range=(0,1200))
```

```python
values_Si_1=np.zeros((ny,nz,100))
for j in range(0,256):
    for k in range(256,512):
        N_Si_II=n_SiII[:,j,k]
        vel_Si=vx[:,j,k]
        values_Si_1[j,k,:],b=np.histogram(vel_Si,bins=100,weights=N_Si_II,range=(0,1200))
```

```python
values_Si_2=np.zeros((ny,nz,100))
for j in range(256,512):
    for k in range(0,256):
        N_Si_II=n_SiII[:,j,k]
        vel_Si=vx[:,j,k]
        values_Si_2[j,k,:],b=np.histogram(vel_Si,bins=100,weights=N_Si_II,range=(0,1200))
```

```python
values_Si_3=np.zeros((ny,nz,100))
for j in range(256,512):
    for k in range(256,512):
        N_Si_II=n_SiII[:,j,k]
        vel_Si=vx[:,j,k]
        values_Si_3[j,k,:],b=np.histogram(vel_Si,bins=100,weights=N_Si_II,range=(0,1200))
```

```python
me=9.10e-28 #in grams
c=3e18 #in Angstroms per sec
dv=12e5  #in cm
dr=4.94e20 #in cm
e = 4.80e-10 #in esu=((cm^3/2)*(grams^1/2))/s
f_SiII=1.18 
f_OVI=0.1325
f_CII=0.128
rw_SiII=1260 # in Angstroms
rw_OVI=1032 #in Angstroms
rw_CII=1335 #in Angstroms
```

```python
Tau_jk_SiII= ((((np.pi)*(e**2))/(me*c))*f_SiII*values_Si*rw_SiII*(dr/dv))
Tau_jk_SiII_1= ((((np.pi)*(e**2))/(me*c))*f_SiII*values_Si_1*rw_SiII*(dr/dv))
Tau_jk_SiII_2= ((((np.pi)*(e**2))/(me*c))*f_SiII*values_Si_2*rw_SiII*(dr/dv))
Tau_jk_SiII_3= ((((np.pi)*(e**2))/(me*c))*f_SiII*values_Si_2*rw_SiII*(dr/dv))
#Tau_jk_OVI= ((((np.pi)*(e**2))/(me*c))*f_OVI*n_v_OVI*rw_OVI*(dr/dv))
#Tau_jk_CII= ((((np.pi)*(e**2))/(me*c))*f_CII*n_v_CII*rw_CII*(dr/dv))
```

```python
T_SiII=Tau_jk_SiII.mean(axis=(0,1))
T_SiII_1=Tau_jk_SiII_1.mean(axis=(0,1))
T_SiII_2=Tau_jk_SiII_2.mean(axis=(0,1))
T_SiII_3=Tau_jk_SiII_3.mean(axis=(0,1))
```

```python
fig = plt.figure(figsize=(16,15))
velocity=np.linspace(0,1200,100)
ax = fig.add_axes([0.19,0.17,0.58,0.75])
plt.plot(velocity, T_SiII,drawstyle='steps', alpha=0.90,linewidth=3.0, color='red',label ='Si II')
#plt.plot(velocity, T_OVI, drawstyle='steps', alpha=0.90,linewidth=4.0, color='black',label ='O VI')
#plt.plot(velocity, new_T_v_SiII, drawstyle='steps', alpha=0.90, color='grey',label ='Si II')
plt.legend(loc='upper right',fontsize=30)
ax.set_xlabel('velocity $km/s$',fontsize=35)
ax.set_ylabel(r'$\tau_{OD}$',fontsize=35)
matplotlib.rc('xtick', labelsize=30) 
matplotlib.rc('ytick', labelsize=30) 
ax.set_xlim(0,800,100)
#plt.savefig(dnameout+'OptDepth_SiII_OVI'+str(i)+'.png', dpi=300)
#plt.close(fig)
plt.show()
```
